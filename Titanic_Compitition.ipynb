```python
from pandas.io.parsers.readers import read_csv
data = pd.read_csv("train.csv")
test =pd.read_csv("test.csv")
test_ids = test["PassengerId"]
```

```python
data.head(5)
```

```python
#cleaning data set
def clean(data):
  data = data.drop(["Name","Ticket","Cabin","PassengerId"], axis=1)
  cols = ["SibSp","Parch","Fare","Age"]
  for col in cols:
    #filling missing values with data
    #using median in place of missing value
    data[col].fillna(data[col].median(), inplace = True)

  #for embarked replacing missing values with unknown token
  data.Embarked.fillna("U",inplace= True)
  return data
#calling clean function for train and test data
data = clean(data)
test = clean(test)
```

```python
data.head(5)
```

```python
#mapping data into intigers
#using sklearn for automapping
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
#lable encoder

cols = ["Sex","Embarked"]

for col in cols:
  #fit and transform on train data
  data[col] = le.fit_transform(data[col])
  #transforming text data
  test[col] = le.transform(test[col])
  print(le.classes_) #jist for test

data.head(5)
```

```python
#model
from sklearn.linear_model import LogisticRegression
#other algoritms can be used like linearregration or knn
from sklearn.model_selection import train_test_split

y = data["Survived"]
x = data.drop("Survived",axis = 1)

x_train,x_val,y_train,y_val = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#classifier
clf = LogisticRegression(random_state=0,max_iter=1000).fit(x_train,y_train)
```

```python
#prediction on validation data
predictions = clf.predict(x_val)
#to get accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_val,predictions)
```

```python
#submission predictions
submission_preds = clf.predict(test)
```

```python
#creating a csv file
#need a dataframe
df = pd.DataFrame({"PassengerId":test_ids.values,
                   "Survived":submission_preds})
```

```python
df.to_csv("submission.csv",index = False)
```
